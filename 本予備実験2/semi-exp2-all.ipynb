```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib

# Load the uploaded CSV file to examine its structure
file_path = '実験２：聞き分けの限界（回答） - フォームの回答 1 (3).csv'
data = pd.read_csv(file_path)
import matplotlib.pyplot as plt

# Extract the experiment scores and order data
experiment_order_columns = [col for col in data.columns if '実験の順番' in col]
experiment_score_columns = [col for col in data.columns if 'スコア' in col]

# Prepare the data for plotting
plot_data = []

for order_col, score_col in zip(experiment_order_columns, experiment_score_columns):
    order = data[order_col]
    scores = data[score_col]
    for ord_value, score in zip(order, scores):
        plot_data.append((ord_value, score))

# Convert to DataFrame for easier plotting
plot_df = pd.DataFrame(plot_data, columns=["Experiment Order", "Score"])

# Create the boxplot
plt.figure(figsize=(12, 6))
plot_df.boxplot(column="Score", by="Experiment Order", grid=False, rot=45)
plt.title("実験２：各手法ごとの正解数")
plt.suptitle("")  # Suppress the automatic title to clean up the plot
plt.xlabel("Method")
plt.ylabel("Score")
plt.show()
```

```python
# ラベルのマッピングを作成
order_labels = {
    "(3,0,0)": "Frequency3段階",
    "(5,0,0)": "Frequency5段階",
    "(0,3,0)": "Amplitude3段階",
    "(0,5,0)": "Amplitude5段階",
    "(0,0,3)": "Pan3段階",
    "(0,0,5)": "Pan5段階",
    "(3,3,3)": "All3段階",
    "(5,5,5)": "All5段階",
}

# Experiment Order列の値をラベルに置き換え
plot_df["Experiment Order"] = plot_df["Experiment Order"].replace(order_labels)

# Boxplotを再度描画
plt.figure(figsize=(12, 6))
plot_df.boxplot(column="Score", by="Experiment Order", grid=False, rot=45)
# plt.title("実験２：各手法ごとの正解数")
plt.suptitle("")  # Suppress the automatic title to clean up the plot
plt.xlabel("Condition")
plt.ylabel("Score")
plt.show()
```
