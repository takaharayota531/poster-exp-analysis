```python
import pandas as pd

# ファイルを読み込む
file_path = 'スコア表 - シート1 (1).csv'
data = pd.read_csv(file_path)

# データの先頭を確認
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib

# 実験名と正答率に関する列を選択
experiments = ["実験１",  "実験3"]

for exp in experiments:
    data_cleaned = []

    if f"{exp}音あり正答率" in data.columns and f"{exp}音無し正答率" in data.columns:
        for condition, label in [("音あり", f"{exp}音あり正答率"), ("音無し", f"{exp}音無し正答率")]:
            data_cleaned.append(pd.DataFrame({
                "Condition": [condition] * len(data),
                "Experiment": [exp] * len(data),
                "Accuracy": data[label]
            }))

    # すべてを結合
    data_combined = pd.concat(data_cleaned, ignore_index=True)

    # 箱ひげ図をプロット
    plt.figure(figsize=(10, 6))
    sns.boxplot(x="Experiment", y="Accuracy", hue="Condition", data=data_combined, palette={"音あり": "blue", "音無し": "orange"})
    plt.title("実験ごとの音あり・音無し正答率の比較")
    plt.ylabel("正答率")
    plt.ylim(0, 1.2)
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.legend(title="Condition", loc="lower right")
    plt.tight_layout()
    plt.show()
```
