```python
import pandas as pd

# ファイルを読み込む
file_path = 'スコア表 - シート1 (1).csv'
data = pd.read_csv(file_path)

# データの先頭を確認
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib

# 実験名と正答率に関する列を選択
experiments = ["実験１",  "実験3"]

for exp in experiments:
    data_cleaned = []

    if f"{exp}音あり正答率" in data.columns and f"{exp}音無し正答率" in data.columns:
        for condition, label in [("Sound", f"{exp}音あり正答率"), ("No Sound", f"{exp}音無し正答率")]:
            data_cleaned.append(pd.DataFrame({
                "Condition": [condition] * len(data),
                "Experiment": [exp] * len(data),
                "Accuracy": data[label]*100
            }))

    # すべてを結合
    data_combined = pd.concat(data_cleaned, ignore_index=True)
    plt.rcParams.update({'font.size': 20})
    # 箱ひげ図をプロット
    plt.figure(figsize=(10, 6))
    sns.boxplot(x="Condition", y="Accuracy",  data=data_combined, boxprops=dict(facecolor='none'))## palette={"音あり": "blue", "音無し": "orange"})
    plt.title("")
    plt.ylabel("Acquisition Accuracy (%) ")
    plt.ylim(0, 110)
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    # plt.legend(title="Condition", loc="lower right")
    plt.tight_layout()
    plt.show()
```
