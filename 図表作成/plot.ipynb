```python
import matplotlib.pyplot as plt
import numpy as np

# 
distance = np.linspace(-0.1, 0.1, 10000)
print(distance)
# 
discreteZ = np.floor(distance / 0.02) * 0.02
print(discreteZ)
amplitude_discrete_exponential = np.power(1.5, -discreteZ / 0.01)
amplitude_continuous_exponential = np.power(1.5, -distance * 100)
amplitude_continuous_linear =  (1 - 12 * distance)
amplitude_discrete_linear =  (1 - 12 * discreteZ)

# 
fig, ax = plt.subplots(figsize=(10, 10))

# 
# target_sphere = plt.Circle((0, 1.0), 0.05, color='blue', alpha=0.2, zorder=0)
# ax.add_artist(target_sphere)

# 
ax.plot(distance, amplitude_discrete_exponential, label='Discrete Exponential', color='tab:blue', zorder=1)
ax.plot(distance, amplitude_continuous_exponential, label='Continuous Exponential', color='tab:orange', zorder=1)
ax.plot(distance, amplitude_continuous_linear, label='Continuous Linear', color='tab:green', zorder=1)
ax.plot(distance, amplitude_discrete_linear, label='Discrete Linear', color='tab:red', zorder=1)

# 
ax.set_xlabel('Distance (m)')
ax.set_ylabel('Amplitude')
ax.set_xlim(-0.05, 0.05)
ax.set_ylim(0, 3.0)
ax.set_aspect('auto')
ax.plot(0.5, 0.5, 'o', color='black', zorder=2)  # 

# 
fig.suptitle('Change in Amplitude with Distance for Different Methods')
ax.legend()

# 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 
plt.rcParams.update({'font.size': 20})

# 
distance = np.linspace(-0.1, 0.1, 10000)

# 
discreteZ = np.floor(distance / 0.01) * 0.01
amplitude_discrete_exponential = np.power(1.5, -discreteZ / 0.01)
amplitude_continuous_exponential = np.power(1.5, -distance * 100)
amplitude_continuous_linear = (1 - 12 * distance)
amplitude_discrete_linear = (1 - 12 * discreteZ)

# 
fig, axs = plt.subplots(2, 2, figsize=(15, 15))

# 
# 
axs[0, 0].plot(distance*100, amplitude_discrete_exponential, label='Discrete Exponential', color='tab:blue')
axs[0, 0].plot(distance*100, amplitude_continuous_exponential, color='tab:orange',alpha=0.3)
axs[0, 0].set_title('Discrete Exponential')
axs[0, 0].set_xlabel('diffX or diffY or diffZ (cm)')
axs[0, 0].set_ylabel('Amplitude or Frequency or Pan')
axs[0, 0].set_xlim(-5, 5)
axs[0, 0].set_ylim(0, 3.0)
axs[0, 0].yaxis.set_ticks([])


# 
axs[0, 1].plot(distance*100, amplitude_continuous_exponential, label='Continuous Exponential', color='tab:orange')
axs[0, 1].set_title('Continuous Exponential')
axs[0, 1].set_xlabel('diffX or diffY or diffZ (cm)')
axs[0, 1].set_ylabel('Amplitude or Frequency or Pan')
axs[0, 1].set_xlim(-5, 5)
axs[0, 1].set_ylim(0, 3.0)
axs[0, 1].yaxis.set_ticks([])

# 
axs[1, 0].plot(distance*100, amplitude_continuous_linear, label='Continuous Linear', color='tab:green')
axs[1, 0].set_title('Continuous Linear')
axs[1, 0].set_xlabel('diffX or diffY or diffZ (cm)')
axs[1, 0].set_ylabel('Amplitude or Frequency or Pan')
axs[1, 0].set_xlim(-5, 5)
axs[1, 0].set_ylim(0, 3.0)
axs[1, 0].yaxis.set_ticks([])

# 
axs[1, 1].plot(distance*100, amplitude_discrete_linear, label='Discrete Linear', color='tab:red')
axs[1, 1].plot(distance*100, amplitude_continuous_linear, color='tab:green',alpha=0.3)
axs[1, 1].set_title('Discrete Linear')
axs[1, 1].set_xlabel('diffX or diffY or diffZ (cm)')
axs[1, 1].set_ylabel('Amplitude or Frequency or Pan')
axs[1, 1].set_xlim(-5, 5)
axs[1, 1].set_ylim(0, 3.0)
axs[1, 1].yaxis.set_ticks([])

# 
plt.tight_layout()
plt.show()
```
