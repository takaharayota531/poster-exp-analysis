```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# 1) データ作成
# -----------------------------
# x軸は -2 から +1 まで
x = np.linspace(-2, 1, 200)

# 連続的な指数関数 y = e^(-x) を例に
y_cont = np.exp(-x)

# 離散ステップ用のxとy (各区間ごとに一定値になるように)
# 例: xステップを -2, -1, 0, 1 の4点
x_step_points = np.array([-2, -1, 0, 1])
y_step_points = np.exp(-x_step_points)  # 離散的な値

# -----------------------------
# 2) プロット
# -----------------------------
plt.figure(figsize=(6,5))

# (A) 連続的な指数関数をオレンジ色で描画
plt.plot(x, y_cont, color='orange', label='Continuous exponential')

# (B) 離散ステップ状のグラフを青色で描画
#     where='post' は「区間の右端」まで一定値にするモード
plt.step(x_step_points, y_step_points, where='post', 
         color='blue', label='Discrete steps')

# (C) ターゲット物体の挿入例 (赤い円マーカー＋ラベル)
target_x, target_y = 0.3, 0.8
plt.plot(target_x, target_y, 'ro', markersize=8)  # 散布点
plt.text(target_x+0.05, target_y, 'Target Object', 
         color='red', fontsize=11, va='center')

# -----------------------------
# 3) 軸ラベルやタイトルなどの装飾
# -----------------------------
plt.title('Discrete Exponential', fontsize=16)
plt.xlabel('x or y or z (cm)', fontsize=12)
plt.ylabel('Amplitude or Frequency or Pan', fontsize=12)

# 凡例と範囲調整
plt.legend()
plt.xlim(-2.2, 1.2)
plt.ylim(0, 8)  # お好みで上下限を調整してください

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# -----------------------------
# 1) データ作成
# -----------------------------
# x軸は -2 から +1 まで
x = np.linspace(-3, 3, 200)

# 連続的な指数関数 y = e^(-x) を例に
y_cont = 2**(-x-2)+4

# 離散ステップ用の x と y (各区間ごとに一定値になるように)
# x_step_points = np.array([-2, -1, 0, 1])
# y_step_points = 2**(-x_step_points)  # 離散的な値
x_step_points = np.linspace(-3, 3, num=4)  # -2から1まで10個のステップ
y_step_points = 2**(-x_step_points-2)+4

# -----------------------------
# 2) プロット
# -----------------------------
plt.figure(figsize=(4,5))

# (A) 連続的な指数関数をオレンジ色で描画
plt.plot(x, y_cont, color='orange', label='Continuous Exponential')

# (B) 離散ステップ状のグラフを青色で描画
plt.step(x_step_points, y_step_points, where='post', 
         color='blue', label='Discrete Exponential')

# -----------------------------
# 3) ターゲット物体 (四角形) の挿入
# -----------------------------
# Rectangle( (左下のx座標, 左下のy座標), 幅, 高さ )
target_rect = patches.Rectangle(
    xy=(-1, 4),      # (x, y): 四角形の左下の位置
    width=2,          # 幅
    height=1.0,         # 高さ
    facecolor='red',    # 塗り色
    alpha=0.2,          # 透明度
    edgecolor='red',    # 枠線の色
    linewidth=1.5
)
# 現在のAxesに追加
plt.gca().add_patch(target_rect)

# 四角形にラベルを付ける
plt.text(-0.5, 5.2, 'Target Object', color='red', fontsize=13,)

# -----------------------------
# 4) 軸ラベルやタイトルなどの装飾
# -----------------------------
plt.title('Discrete Exponential', fontsize=16)
plt.xlabel('Y', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.legend()
plt.xlim(-3.2, 3.2)
plt.ylim(2.5, 6.5)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# -----------------------------
# 1) データ作成
# -----------------------------
# x軸は -4 から +4 まで
x = np.linspace(-4, 4, 200)

# 連続的な指数関数 y = 2^(-x)
y_cont_exp = 2**(-x)
# 離散ステップ状の指数関数
discrete_x_exp = np.linspace(-4, 4, num=9)
discrete_y_exp = 2**(-discrete_x_exp)

# 連続的な線形関数 y = -x + 4
y_cont_lin = -x + 4
# 離散ステップ状の線形関数
discrete_x_lin = np.linspace(-4, 4, num=9)
discrete_y_lin = -discrete_x_lin + 4

# -----------------------------
# 2) プロット
# -----------------------------
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# (A) 離散指数関数
axs[0, 0].plot(x, y_cont_exp, color='orange', label='Continuous Exponential')
axs[0, 0].step(discrete_x_exp, discrete_y_exp, where='post', color='blue', label='Discrete Exponential')
axs[0, 0].set_title('Discrete Exponential', fontsize=14)
axs[0, 0].legend()

# (B) 連続指数関数
axs[0, 1].plot(x, y_cont_exp, color='orange', label='Continuous Exponential')
axs[0, 1].set_title('Continuous Exponential', fontsize=14)
axs[0, 1].legend()

# (C) 連続線形関数
axs[1, 0].plot(x, y_cont_lin, color='green', label='Continuous Linear')
axs[1, 0].set_title('Continuous Linear', fontsize=14)
axs[1, 0].legend()

# (D) 離散線形関数
axs[1, 1].plot(x, y_cont_lin, color='green', alpha=0.4, label='Continuous Linear')
axs[1, 1].step(discrete_x_lin, discrete_y_lin, where='post', color='red', label='Discrete Linear')
axs[1, 1].set_title('Discrete Linear', fontsize=14)
axs[1, 1].legend()

# -----------------------------
# 3) 四角形の挿入
# -----------------------------
target_rect = patches.Rectangle(
    xy=(-2, 2),      # (x, y): 四角形の左下の位置
    width=4,         # 幅
    height=1.0,      # 高さ
    facecolor='red', # 塗り色
    alpha=0.2,       # 透明度
    edgecolor='red', # 枠線の色
    linewidth=1.5
)
for ax in axs.flat:
    ax.add_patch(target_rect)
    ax.text(-1.8, 3.2, 'Target Object', color='red', fontsize=10)

# -----------------------------
# 4) 軸ラベルや装飾
# -----------------------------
for ax in axs.flat:
    ax.set_xlabel('diffX or diffY or diffZ (cm)', fontsize=10)
    ax.set_ylabel('Amplitude of Frequency or Pan', fontsize=10)
    ax.set_xlim(-4.2, 4.2)
    ax.set_ylim(0, 8)
    ax.legend()

plt.tight_layout()
plt.show()
```
